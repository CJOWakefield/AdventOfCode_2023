```python
def is_valid(map, groups):
    pos = 0
    seen = []
    for char in map:
        if char == '.':
            if pos > 0:
                seen.append(pos)
            pos = 0
        elif char == '#':
            pos += 1
        else:
            assert False
    if pos > 0:
        seen.append(pos)
    return seen == groups


def combinations(map, groups, position):
    if position == len(map):
        return 1 if is_valid(map, groups) else 0
    if map[position] == '?':
        return (combinations(map[:position] + '#' + map[position+1:], groups, position + 1) + 
                combinations(map[:position] + '.' + map[position+1:], groups, position + 1))
    else:
        return combinations(map, groups, position+1)

score = 0

for row in input:
    map, groups = row.split()
    groups = [int(num) for num in groups.split(',')]
    score_curr = combinations(map, groups, 0)
    score += score_curr
    
score
```

```python
from functools import cache

@cache
def recurse(lava, springs, result=0):
    if not springs:
        return '#' not in lava
    current, springs = springs[0], springs[1:]
    for i in range(len(lava) - sum(springs) - len(springs) - current + 1):
        if "#" in lava[:i]:
            break
        if (nxt := i + current) <= len(lava) and '.' not in lava[i : nxt] and lava[nxt : nxt + 1] != "#":
            result += recurse(lava[nxt + 1:], springs)
    return result


data = [x.split() for x in input]
p1, p2 = 0, 0
for e, (lava, springs) in enumerate(data):
    p1 += recurse(lava, (springs := tuple(map(int, springs.split(",")))))
    p2 += recurse("?".join([lava] * 5), springs * 5)
print(p1, p2)
```
