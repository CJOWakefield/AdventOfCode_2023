```python
from itertools import groupby

input = [x for x in open("13.txt").read().strip().split('\n')]
clean_input = [list(g) for k, g in groupby(input, key=bool) if k]
clean_input
```

```python
def reflection_sum(map):
    reflection_index = 0
    for row_i, row in enumerate(map):
        if row_i < len(map)-1:
            if row == map[row_i + 1]:
                reflection_index = row_i
                overlap_length = min(abs(0-row_i), abs(len(map)-row_i-2))
                if map[row_i-overlap_length:row_i] == map[row_i+2: row_i+2+overlap_length]:
                    return len(map[0:row_i+1])
        
    return 0
```

```python
def reflection_sum(map):
    
    for row_i, row in enumerate(map):
        if row_i < len(map)-1:
            section = min(len(map) - row_i, row_i)
            top = map[row_i - section: row_i]
            top_flipped = top[::-1]
            bottom = map[row_i: row_i + section]
            if top_flipped == bottom:
                return row_i
                
    return 0
```

```python
result = 0

for map in clean_input:
    horizontal = map
    vertical = [''.join([char for char in row]) for row in list(zip(*map]))]

    map_result = (reflection_sum(horizontal) * 100) + reflection_sum(vertical)
    result += map_result
    print(map_result)
```

```python
import numpy as np

puzzles = [i.split('\n') for i in open('13.txt').read().split('\n\n')]

for p in range(0 , len(puzzles)):
    for l in range(0, len(puzzles[p])):
        puzzles[p][l] = puzzles[p][l].replace('.', '0')
        puzzles[p][l] = puzzles[p][l].replace('#', '1')
        puzzles[p][l] = [int(x) for x in puzzles[p][l]]
    puzzles[p] = np.array(puzzles[p])

def findMirror(p, multiplier, part):
    value = 0
    for l in range(1, p.shape[0]):
        block_size = min(p.shape[0] - l, l)
        top_segment = p[l-block_size:l]
        remaining_bot_segment = p[l:l+block_size]
        top_segment_mirrored = np.flip(top_segment, axis=0)
        if part == 1:
            if np.all(top_segment_mirrored == remaining_bot_segment):
                value = l * multiplier
        if part == 2:
            if list((top_segment_mirrored == remaining_bot_segment).flatten()).count(False) == 1:
                value = l * multiplier
    return value

parts = [1,2]
for part in parts:
    point = 0
    for p in puzzles:
        point += findMirror(p, 100, part)
        point += findMirror(np.rot90(p,-1), 1, part)
    print("part: ", part, point)
```
