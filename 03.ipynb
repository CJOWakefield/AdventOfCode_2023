```python
def valid_character(input, row_i, row, char_i):
    
    adj_squares = [input[row_i][max([char_i-1, 0])],
                   input[row_i][min([char_i+1, len(row)-1])]]

    if row_i < len(input)-1:
        adj_squares.extend([input[row_i+1][max([char_i-1, 0])],
                           input[row_i+1][char_i],
                           input[row_i+1][min([char_i+1, len(row)-1])]])
    
    if row_i > 0:
        adj_squares.extend([input[row_i-1][max([char_i-1, 0])],
                            input[row_i-1][char_i],
                            input[row_i-1][min([char_i+1, len(row)-1])]])

    return any(check in adj_squares for check in symbols)

rolling_count = 0
number_values = []
is_valid = False

for row_i, row in enumerate(input):
    symbols = [i for i in string.punctuation.replace('.', '')]
    for char_i, char in enumerate(row):
        if char in string.digits:
            number_values.append(char)
            if valid_character(input, row_i, row, char_i):
                is_valid = True
                
        else:
            if is_valid:
                result = int(''.join(number_values))
                rolling_count += result
            number_values.clear()
            is_valid = False
            
rolling_count
                
```
