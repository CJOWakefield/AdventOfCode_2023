```python
data = open('15.txt').read().split(',')

def HASH(string):
    total = 0
    for char in string:
        total += ord(char)
        total = (total * 17) % 256
    return total

sum([HASH(str) for str in data])
```

# Part 2

```python
box_dict = {}

for str in data:
    if '=' in str:
        str = str.split('=')
        hash = HASH(str[0])
        if hash not in box_dict.keys():
            box_dict[hash] = []
            box_dict[hash].append([str[0], str[1]])
        elif str[0] in [val[0] for val in box_dict[hash]]:
            curr_list = [val[0] for val in box_dict[hash]]
            idx = curr_list.index(str[0])
            box_dict[hash][idx] = [str[0], str[1]]
        else:
            box_dict[hash].append([str[0], str[1]])
    else:
        str = str.split('-')
        hash = HASH(str[0])
        if hash in box_dict.keys() and str[0] in [val[0] for val in box_dict[hash]]:
            curr_list = [val[0] for val in box_dict[hash]]
            idx = curr_list.index(str[0])
            del box_dict[hash][idx]

myKeys = list(box_dict.keys())
myKeys.sort()
box_dict = {i: box_dict[i] for i in myKeys}

box_dict
```

```python
final = 0

for key in box_dict.keys():
    key_total = 0
    if box_dict[key]:
        for lens_i, lens in enumerate(box_dict[key]):
            key_total += int(key + 1) * (lens_i + 1) * int(lens[1])
    final += key_total
```
