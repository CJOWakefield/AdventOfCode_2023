```python
import pandas as pd

grid = input
grid_expanded = []

for row in grid:
    grid_expanded.append(row)
    if '#' not in row:
        grid_expanded.append(row)

grid = list(zip(*grid_expanded))
grid_final = []

for row in grid:
    grid_final.append(row)
    if '#' not in row:
        grid_final.append(row)

num_count = 0
num_positions = {}

for row_i, row in enumerate(grid_final):
    for char_i, char in enumerate(row):
        if char == '#':
            num_positions[num_count] = [row_i, char_i]
            num_count += 1
```

```python
final_distance_grid = []

for key in num_positions.keys():
    distances = []
    for j in range(key+1, len(num_positions.keys())):
        row_diff = abs(num_positions[j][0] - num_positions[key][0])
        column_diff = abs(num_positions[j][1] - num_positions[key][1])
        min_step_distance = row_diff + column_diff
        distances.append(min_step_distance)
    final_distance_grid.append(sum(distances))
    
sum(final_distance_grid)
```

```python
num_positions_base = {}
num_count_base = 0

for row_i, row in enumerate(input):
    for char_i, char in enumerate(row):
        if char == '#':
            num_positions_2[num_count_2] = [row_i, char_i]
            num_count_2 += 1

non_empty_rows = [num[0] for num in num_positions_2.values()]
non_empty_cols = [num[1] for num in num_positions_2.values()]

empty_rows = [row for row in range(140) if row not in non_empty_rows]
empty_cols = [col for col in range(140) if col not in non_empty_cols]

positions_final = {}
num_count_final = 0
ef = 1000000 - 1

for row_i, row in enumerate(input):
    for char_i, char in enumerate(row):
        if char == '#':
            prev_empty_rows = len([row for row in empty_rows if row < row_i])
            prev_empty_cols = len([col for col in empty_cols if col < char_i])
            positions_final[num_count_final] = [row_i + (prev_empty_rows * ef), char_i + (prev_empty_cols * ef)]
            num_count_final += 1

dist_grid = []

for key in positions_final.keys():
    distances = []
    for j in range(key+1, len(positions_final.keys())):
        row_diff = abs(positions_final[j][0] - positions_final[key][0])
        column_diff = abs(positions_final[j][1] - positions_final[key][1])
        min_step_distance = row_diff + column_diff
        distances.append(min_step_distance)
    dist_grid.append(sum(distances))
    
sum(dist_grid)          
```
