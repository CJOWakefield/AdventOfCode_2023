```python
def data_mapping(data):

    cleaned_data = data.copy()

    for row_i, row in enumerate(cleaned_data):
        if ':' in row:
            cleaned_data[row_i] = row.split(':')[1]
        else:
            cleaned_data[row_i] = row.split()
    cleaned_data[0] = cleaned_data[0].strip().split()

    mapping_dict = {'map_1': []}
    
    map_value = True
    map_i = 1
    
    for map_pos, map in enumerate(cleaned_data):
        if map:
            mapping_dict[f'map_{map_i}'] += map
        elif not map and cleaned_data[map_pos+1]:
            map_i += 1
            mapping_dict[f'map_{map_i}'] = []

    for mapping in mapping_dict.keys():
        if mapping != 'map_1':
            map = mapping_dict[mapping]
            new_mapping = []
            current_trio = []
            tally = 1
            for map_i, map_value in enumerate(map):
                if not tally % 3 == 0:
                    current_trio.append(int(map_value))
                    tally += 1
                else:
                    current_trio.append(int(map_value))
                    new_mapping.append(current_trio)
                    current_trio = []
                    tally = 1
            mapping_dict[mapping] = new_mapping
        else:
            map = [int(value) for value in mapping_dict[mapping]]
            mapping_dict[mapping] = map
            
    return mapping_dict
```

```python
def map_translation(input_seed, map):
    current_mapping = mapping[map]
    translation = input_seed
    for trio in current_mapping:
        if input_seed in range(trio[1], trio[1]+trio[2]):
            translation = (translation - trio[1]) + trio[0]
            break
    return translation
```

```python
from functools import reduce

seeds, *mappings = open('05_input.txt').read().split('\n\n')
seeds = list(map(int, seeds.split()[1:]))

def lookup(inputs, mapping):
    for start, length in inputs:
        while length > 0:
            for m in mapping.split('\n')[1:]:
                dst, src, len = map(int, m.split())
                delta = start - src
                if delta in range(len):
                    len = min(len - delta, length)
                    yield (dst + delta, len)
                    start += len
                    length -= len
                    break
            else: yield (start, length); break

print(*[min(reduce(lookup, mappings, s))[0] for s in [
    zip(seeds, [1] * len(seeds)),
    zip(seeds[0::2], seeds[1::2])]])
```
